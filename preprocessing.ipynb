## song part indicator removal

```python
import re
def remove_feature(text):
    return re.sub(r'\[.*?\]', '', text).strip()

df['lyrics_cleaned'] = df['lyrics'].apply(remove_feature)

for text in df['lyrics_cleaned'][3374193:]:
    print(text)
```

## feature cleaned

```python
df['features'][6]
```

```python
import re
def feature_cleaned(text):
    cleaned_text = re.sub(r'[{}\'"]', '', text)  
    cleaned_text = re.sub(r',', ' ', cleaned_text) 
    return cleaned_text.strip()

df['feature_cleaned'] = df['features'].apply(feature_cleaned)

for text in df['feature_cleaned'][3374193:]:
    print(text)
```

```python
df['song_document'] = df['title'] + " " + df['lyrics_cleaned'] + " " + df['feature_cleaned']

for text in df['song_document'][3374193:]:
    print(text)
```
