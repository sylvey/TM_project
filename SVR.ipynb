```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the dataset
df = pd.read_csv("data/song_lyrics_en.csv")  # Adjust the path as necessary

# Limit the dataset to the first 10,000 rows for training
df_limited = df.head(10000)

# Text encoding: Bag of Words
count_vectorizer = CountVectorizer(max_features=5000)  # Limit to top 5000 features
X_bow = count_vectorizer.fit_transform(df_limited['lyrics']).toarray()

# Text encoding: TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf_vectorizer.fit_transform(df_limited['lyrics']).toarray()

# Target variable (assuming 'views' column is the popularity measure)
y = df_limited['views']

# Split the data into training and testing sets (80% train, 20% test)
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=42)
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Train an SVR model using Bag of Words features
svr_bow = SVR(kernel='linear')
svr_bow.fit(X_train_bow, y_train)
y_pred_bow = svr_bow.predict(X_test_bow)

# Train an SVR model using TF-IDF features
svr_tfidf = SVR(kernel='linear')
svr_tfidf.fit(X_train_tfidf, y_train)
y_pred_tfidf = svr_tfidf.predict(X_test_tfidf)

# Evaluate the models
mse_bow = mean_squared_error(y_test, y_pred_bow)
mse_tfidf = mean_squared_error(y_test, y_pred_tfidf)

print(f'Mean Squared Error (Bag of Words): {mse_bow}')
print(f'Mean Squared Error (TF-IDF): {mse_tfidf}')
```
