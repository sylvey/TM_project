```python
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf

# 檢查是否有可用的 GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# 加載預訓練的 BERT 模型和 tokenizer
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 示例文本數據 (實際上是你的文本數據集)
texts = df['song_document'].tolist()

import numpy as np

batch_size = 32  # 調整這個數字來適應可用的 GPU 記憶體
num_batches = int(np.ceil(len(texts) / batch_size))

all_cls_embeddings = []

for i in range(num_batches):
    batch_texts = texts[i * batch_size:(i + 1) * batch_size]
    encoded_inputs = tokenizer(batch_texts, return_tensors='tf', padding=True, truncation=True, max_length=128)

    # 將數據移動到 GPU 上，並處理該批次
    with tf.device('/GPU:0'):
        bert_outputs = bert_model(encoded_inputs)

    # 提取 [CLS] token 的嵌入
    cls_embeddings = bert_outputs.last_hidden_state[:, 0, :]
    all_cls_embeddings.append(cls_embeddings)

# 將所有批次的結果合併
all_cls_embeddings = tf.concat(all_cls_embeddings, axis=0)
```

```python
import numpy as np


# Convert the other features to TensorFlow tensors
artist_embeddings_tf = tf.convert_to_tensor(artist_embeddings, dtype=tf.float32)
tag_embeddings_tf = tf.convert_to_tensor(tag_embeddings, dtype=tf.float32)

# Concatenate the BERT embeddings with the other feature embeddings
combined_embeddings = tf.concat([all_cls_embeddings, tag_embeddings_tf, artist_embeddings_tf], axis=1)

# Combined shape: (batch_size, hidden_size + other_feature_size)
print(combined_embeddings.shape)
```

```python
def transform_y(label):
    if label == 'Very Low':
        return 0
    if label == 'Low':
        return 1
    if label == 'Medium':
        return 2
    if label == 'High':
        return 3
    if label == 'Very High':
        return 4

y = df['popularity_bin'].apply(transform_y).values
```

```python
y.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# Assuming X and y are your input and output arrays
# X.shape = (10000, 903), y.shape = (10000,)

# Convert y to one-hot encoded format for multi-class classification
y_one_hot = to_categorical(y, num_classes=5)
y_one_hot
```

```python
X_train, X_test, y_train, y_test = train_test_split(combined_embeddings.numpy(), y_one_hot, test_size=0.1, random_state=42)
```

```python
y_train.shape
```

```python
from tensorflow.keras import backend as K

def f1_score(y_true, y_pred):
    y_pred_classes = K.argmax(y_pred, axis=-1)
    y_true_classes = K.argmax(y_true, axis=-1)

    precision = tf.metrics.Precision()(y_true_classes, y_pred_classes)
    recall = tf.metrics.Recall()(y_true_classes, y_pred_classes)

    f1 = 2 * ((precision * recall) / (precision + recall + K.epsilon()))
    return f1
```

```python
model1 = Sequential()

# Input layer and first hidden layer (Dense)
model1.add(Dense(128, input_dim=1371, activation='relu'))  # Input dim matches the number of features in X

# Add additional hidden layers as needed
model1.add(Dense(64, activation='relu'))

# Output layer (for 5 classes)
model1.add(Dense(5, activation='softmax'))  # Softmax for multi-class classification


model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model1.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

loss, accuracy = model1.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import f1_score
import numpy as np

# Get predictions from the model (these will be probabilities)
y_pred_prob = model1.predict(X_test)

# Convert probabilities to class predictions (e.g., the index of the max value)
y_pred = np.argmax(y_pred_prob, axis=1)

# If y_test is one-hot encoded, convert it back to class labels
y_test_class = np.argmax(y_test, axis=1)

# Calculate the F1 score using sklearn
f1 = f1_score(y_test_class, y_pred, average='weighted')

# Print the F1 score
print(f"Test F1 Score: {f1:.2f}")
```
